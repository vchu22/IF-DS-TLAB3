```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import coffee data
coffee = pd.read_csv("../data/crop/coffee_output.csv")

# import weather data
weather = pd.read_csv("../data/weather/weater_output.csv")
```

```python
# TODO: Merge these two dataframes together

coffee_weather = pd.merge(coffee, weather, left_on='year', right_on='year')
```

```python
# TODO: View the first 5 rows of your data

coffee_weather.head()
```

```python
# TODO: Check for null values 

coffee_weather.isnull().sum()
```

```python
# TODO: Drop rows that contain null values

coffee_weather_cleaned = coffee_weather.dropna()
coffee_weather_cleaned
```

```python
# TODO: BEGIN YOUR EDA!

# Q1, calculate the Pearson's Correlation Coefficient between temp_avg and 60kgs_bag
sns.scatterplot(data=coffee_weather_cleaned, x="temp_avg", y="60kgs_bag")
coffee_weather_cleaned[["temp_avg", "60kgs_bag"]].corr()
```
